```python
###Calc Daily VOL_MA_3###

ddb_wdb_ref["VOL_MA3"] = ddb_wdb_ref.groupby(ddb_wdb_ref["Name"]).Volume.transform(lambda c: c.rolling(3).mean())
ddb_wdb_ref["VOL_MA3"] = ddb_wdb_ref["VOL_MA3"].fillna(0)

###Calc Daily VOL EMA_3###


ddb_wdb_ref.loc[(ddb_wdb_ref['#']<3), 'Vol_EMA3'] =  0
ddb_wdb_ref.loc[(ddb_wdb_ref['#']==3), 'Vol_EMA3'] =  ddb_wdb_ref["Avg_Vol"]
ddb_wdb_ref.loc[(ddb_wdb_ref['#']>3), 'Vol_EMA3'] = ddb_wdb_ref.groupby(ddb_wdb_ref["Name"]).Volume.transform(lambda c: c.ewm(span=3, adjust=False).mean())
    

```

```python
######## SELL WHEN BELOW EMA8 or if profit above X% SELL BELOW EMA3 AND VOL BIGGER THEN AVG VOL BY X%##########

if D_Sell_EMA8_exe==1:
    if D_Sell_EMA3_exe==1:
        if D_VOL_EMA3_exe==1:
            
            ddb_wdb_ref["Symbol_Pos"]=""

            count = 0

            while count < x:

                #Sell_VOL_EMA3#
                
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] == 1), 'Sell_VOL_EMA3'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref['Volume']*(1+VOL_EMA3_Sell_Above)>=ddb_wdb_ref['Vol_EMA3']), 'Sell_VOL_EMA3'] = 1
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref['Volume']*(1+VOL_EMA3_Sell_Above)<ddb_wdb_ref['Vol_EMA3']), 'Sell_VOL_EMA3'] = 0
                
                 #temp1_VOL_EMA3#
                
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] == 1), 'temp1_VOL_EMA3'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref['BUY_SELL_EMA3']==0)&(ddb_wdb_ref['Sell_VOL_EMA3']==1) & (ddb_wdb_ref['PRICE_EMA3_KICKS']==0) & (ddb_wdb_ref['CLS>EMAJ']==0), 'temp1_VOL_EMA3'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref['BUY_SELL_EMA3']==0)&(ddb_wdb_ref['Sell_VOL_EMA3']==1) & (ddb_wdb_ref['PRICE_EMA3_KICKS']==0) & (ddb_wdb_ref['CLS>EMAJ']==1), 'temp1_VOL_EMA3'] = ddb_wdb_ref['temp1_VOL_EMA3'].shift(1)
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref['BUY_SELL_EMA3']==0)&(ddb_wdb_ref['Sell_VOL_EMA3']==1) & (ddb_wdb_ref['PRICE_EMA3_KICKS']==1) & (ddb_wdb_ref['SELL_EMA3']==1), 'temp1_VOL_EMA3'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref['BUY_SELL_EMA3']==0)&(ddb_wdb_ref['Sell_VOL_EMA3']==1) & (ddb_wdb_ref['PRICE_EMA3_KICKS']==1) & (ddb_wdb_ref['SELL_EMA3']==0), 'temp1_VOL_EMA3'] =  ddb_wdb_ref['temp1_VOL_EMA3'].shift(1)
                
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref['BUY_SELL_EMA3']==0)&(ddb_wdb_ref['Sell_VOL_EMA3']==0) , 'temp1_VOL_EMA3'] = ddb_wdb_ref['temp1_VOL_EMA3'].shift(1)
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref['BUY_SELL_EMA3']==1) , 'temp1_VOL_EMA3'] = ddb_wdb_ref['BUY_SELL_EMA3']
                
                ddb_wdb_ref.loc[(ddb_wdb_ref['PRICE_EMA3_KICKS'] == 1), 'temp2_VOL_EMA3'] = "***"
                
                 ### BUY / SELL / DECITION ###
    
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] == 1), 'Symbol_Pos'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["temp1_VOL_EMA3"]== 1)&(ddb_wdb_ref["temp1_VOL_EMA3"].shift(1)== 0), 'Symbol_Pos'] = 'BUY'
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["temp1_VOL_EMA3"]== 1)&(ddb_wdb_ref["temp1_VOL_EMA3"].shift(1)!= 0), 'Symbol_Pos'] = 'HOLD'
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["temp1_VOL_EMA3"]!= 1)&(ddb_wdb_ref["temp1_VOL_EMA3"].shift(1)!= 1), 'Symbol_Pos'] = 'OUT'
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["temp1_VOL_EMA3"]!= 1)&(ddb_wdb_ref["temp1_VOL_EMA3"].shift(1)== 1), 'Symbol_Pos'] = 'SELL'
                
                ###Adding relevant rows to priceline###
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(ddb_wdb_ref["Symbol_Pos"]!= "SELL"), 'BUY_PRICE_EMA8_LINE'] = ddb_wdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(ddb_wdb_ref["Symbol_Pos"]== "SELL"), 'BUY_PRICE_EMA8_LINE'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(ddb_wdb_ref["Symbol_Pos"]== "OUT"), 'BUY_PRICE_EMA8_LINE'] = 0
                
                #### Calculating Profit ###   
                
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(ddb_wdb_ref["Symbol_Pos"]!= "SELL"), 'PRICE_EMA8_LINE_Profit'] = ((ddb_wdb_ref["Close"]-ddb_wdb_ref["BUY_PRICE_EMA8_LINE"])/ddb_wdb_ref["BUY_PRICE_EMA8_LINE"])
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(ddb_wdb_ref["Symbol_Pos"]== "SELL"), 'PRICE_EMA8_LINE_Profit'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(ddb_wdb_ref["Symbol_Pos"]== "OUT"), 'PRICE_EMA8_LINE_Profit'] = 0

                #### priceline EMA3 Kick ###  
                
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] == 1), 'PRICE_EMA3_KICKS'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["PRICE_EMA8_LINE_Profit"] > EMA_Sell_Above_exe), 'PRICE_EMA3_KICKS'] = 1
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["PRICE_EMA8_LINE_Profit"] <= EMA_Sell_Above_exe)&(ddb_wdb_ref["PRICE_EMA3_KICKS"].shift(1)==1)&(ddb_wdb_ref["temp1_VOL_EMA3"]==1), 'PRICE_EMA3_KICKS'] = 1
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["PRICE_EMA8_LINE_Profit"] <= EMA_Sell_Above_exe)&(ddb_wdb_ref["PRICE_EMA3_KICKS"].shift(1)==1)&(ddb_wdb_ref["temp1_VOL_EMA3"]==0), 'PRICE_EMA3_KICKS'] = 0
                
                ### RESETING WHEN SELL###
          
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["Symbol_Pos"]== "SELL"), 'BUY_PRICE_EMA8_LINE'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["Symbol_Pos"]== "SELL"), 'PRICE_EMA8_LINE_Profit'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["Symbol_Pos"]== "SELL"), 'temp1_VOL_EMA3'] = 0
                ddb_wdb_ref.loc[(ddb_wdb_ref['#'] != 1)&(ddb_wdb_ref["Symbol_Pos"]== "SELL"), 'SELL_EMA3'] = 0

                count = count + 1
```
