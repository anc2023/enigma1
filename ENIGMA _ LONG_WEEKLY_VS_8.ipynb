### Weekly Avg Vol

```python
# AVG Weekly VOLUME START #

wdb_mdb_ref['Accu_Vol'] = wdb_mdb_ref.groupby(wdb_mdb_ref["Name"]).Volume.cumsum()
wdb_mdb_ref['Avg_Vol'] = wdb_mdb_ref['Accu_Vol'] / wdb_mdb_ref["#"]

# AVG VOLUME END #
```

```python
# EMA3 / EMA8 Weekly VOLUME START #

#Weekly_Vol_EMA3#

###Calc Weekly VOL_MA_3###

wdb_mdb_ref["VOL_MA3"] = wdb_mdb_ref.groupby(wdb_mdb_ref["Name"]).Volume.transform(lambda c: c.rolling(3).mean())
wdb_mdb_ref["VOL_MA3"] = wdb_mdb_ref["VOL_MA3"].fillna(0)

###Calc Weekly VOL EMA_3###

wdb_mdb_ref.loc[(wdb_mdb_ref['#']<3), 'Vol_EMA3'] =  0
wdb_mdb_ref.loc[(wdb_mdb_ref['#']==3), 'Vol_EMA3'] =  wdb_mdb_ref["Avg_Vol"]
wdb_mdb_ref.loc[(wdb_mdb_ref['#']>3), 'Vol_EMA3'] = wdb_mdb_ref.groupby(wdb_mdb_ref["Name"]).Volume.transform(lambda c: c.ewm(span=3, adjust=False).mean())
    


#count = 0
    
#while count < xw:
  #  wdb_mdb_ref.loc[(wdb_mdb_ref['#']<3), 'Vol_EMA3'] = 0
  #  wdb_mdb_ref.loc[(wdb_mdb_ref['#']==3), 'Vol_EMA3'] = wdb_mdb_ref["VOL_MA3"]
   # wdb_mdb_ref.loc[(wdb_mdb_ref['#']>3), 'Vol_EMA3'] = ((wdb_mdb_ref["Volume"]*0.5)+(wdb_mdb_ref["Vol_EMA3"].shift(1)*(0.5)))
        
#    count = count + 1
                            

```

```python
######## SELL WHEN BELOW EMA8 or if profit above X% SELL BELOW EMA3 AND VOL BIGGER THEN AVG VOL BY X%##########

if W_Sell_EMA8_exe==1:
    if W_Sell_EMA3_exe==1:
        if W_VOL_EMA3_exe==1:
            
            wdb_mdb_ref["Symbol_Pos"]=""

            count = 0

            while count < x:

                #Sell_VOL_EMA3#
                
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] == 1), 'Sell_VOL_EMA3'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref['Volume']*(1+W_VOL_EMA3_Sell_Above)>=wdb_mdb_ref['Vol_EMA3']), 'Sell_VOL_EMA3'] = 1
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref['Volume']*(1+W_VOL_EMA3_Sell_Above)<wdb_mdb_ref['Vol_EMA3']), 'Sell_VOL_EMA3'] = 0
                
                 #temp1_VOL_EMA3#
                
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] == 1), 'temp1_VOL_EMA3'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref['BUY_SELL_EMA3']==0)&(wdb_mdb_ref['Sell_VOL_EMA3']==1) & (wdb_mdb_ref['PRICE_EMA3_KICKS']==0) & (wdb_mdb_ref['CLS>EMA8']==0), 'temp1_VOL_EMA3'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref['BUY_SELL_EMA3']==0)&(wdb_mdb_ref['Sell_VOL_EMA3']==1) & (wdb_mdb_ref['PRICE_EMA3_KICKS']==0) & (wdb_mdb_ref['CLS>EMA8']==1), 'temp1_VOL_EMA3'] = wdb_mdb_ref['temp1_VOL_EMA3'].shift(1)
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref['BUY_SELL_EMA3']==0)&(wdb_mdb_ref['Sell_VOL_EMA3']==1) & (wdb_mdb_ref['PRICE_EMA3_KICKS']==1) & (wdb_mdb_ref['SELL_EMA3']==1), 'temp1_VOL_EMA3'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref['BUY_SELL_EMA3']==0)&(wdb_mdb_ref['Sell_VOL_EMA3']==1) & (wdb_mdb_ref['PRICE_EMA3_KICKS']==1) & (wdb_mdb_ref['SELL_EMA3']==0), 'temp1_VOL_EMA3'] =  wdb_mdb_ref['temp1_VOL_EMA3'].shift(1)
                
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref['BUY_SELL_EMA3']==0)&(wdb_mdb_ref['Sell_VOL_EMA3']==0) , 'temp1_VOL_EMA3'] = wdb_mdb_ref['temp1_VOL_EMA3'].shift(1)
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref['BUY_SELL_EMA3']==1) , 'temp1_VOL_EMA3'] = wdb_mdb_ref['BUY_SELL_EMA3']
                
                wdb_mdb_ref.loc[(wdb_mdb_ref['PRICE_EMA3_KICKS'] == 1), 'temp2_VOL_EMA3'] = "***"
                
                 ### BUY / SELL / DECITION ###
    
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] == 1), 'Symbol_Pos'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["temp1_VOL_EMA3"]== 1)&(wdb_mdb_ref["temp1_VOL_EMA3"].shift(1)== 0), 'Symbol_Pos'] = 'BUY'
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["temp1_VOL_EMA3"]== 1)&(wdb_mdb_ref["temp1_VOL_EMA3"].shift(1)!= 0), 'Symbol_Pos'] = 'HOLD'
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["temp1_VOL_EMA3"]!= 1)&(wdb_mdb_ref["temp1_VOL_EMA3"].shift(1)!= 1), 'Symbol_Pos'] = 'OUT'
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["temp1_VOL_EMA3"]!= 1)&(wdb_mdb_ref["temp1_VOL_EMA3"].shift(1)== 1), 'Symbol_Pos'] = 'SELL'
                
                ###Adding relevant rows to priceline###
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(wdb_mdb_ref["Symbol_Pos"]!= "SELL"), 'BUY_PRICE_EMA8_LINE'] = wdb_mdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(wdb_mdb_ref["Symbol_Pos"]== "SELL"), 'BUY_PRICE_EMA8_LINE'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(wdb_mdb_ref["Symbol_Pos"]== "OUT"), 'BUY_PRICE_EMA8_LINE'] = 0
                
                #### Calculating Profit ###   
                
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(wdb_mdb_ref["Symbol_Pos"]!= "SELL"), 'PRICE_EMA8_LINE_Profit'] = ((wdb_mdb_ref["Close"]-wdb_mdb_ref["BUY_PRICE_EMA8_LINE"])/wdb_mdb_ref["BUY_PRICE_EMA8_LINE"])
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(wdb_mdb_ref["Symbol_Pos"]== "SELL"), 'PRICE_EMA8_LINE_Profit'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["BUY_PRICE_EMA8_LINE"].shift(1)!= 0)&(wdb_mdb_ref["Symbol_Pos"]== "OUT"), 'PRICE_EMA8_LINE_Profit'] = 0

                #### priceline EMA3 Kick ###  
                
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] == 1), 'PRICE_EMA3_KICKS'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["PRICE_EMA8_LINE_Profit"] > W_EMA_Sell_Above_exe), 'PRICE_EMA3_KICKS'] = 1
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["PRICE_EMA8_LINE_Profit"] <= W_EMA_Sell_Above_exe)&(wdb_mdb_ref["PRICE_EMA3_KICKS"].shift(1)==1)&(wdb_mdb_ref["temp1_VOL_EMA3"]==1), 'PRICE_EMA3_KICKS'] = 1
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["PRICE_EMA8_LINE_Profit"] <= W_EMA_Sell_Above_exe)&(wdb_mdb_ref["PRICE_EMA3_KICKS"].shift(1)==1)&(wdb_mdb_ref["temp1_VOL_EMA3"]==0), 'PRICE_EMA3_KICKS'] = 0
                
                ### RESETING WHEN SELL###
          
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["Symbol_Pos"]== "SELL"), 'BUY_PRICE_EMA8_LINE'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["Symbol_Pos"]== "SELL"), 'PRICE_EMA8_LINE_Profit'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["Symbol_Pos"]== "SELL"), 'temp1_VOL_EMA3'] = 0
                wdb_mdb_ref.loc[(wdb_mdb_ref['#'] != 1)&(wdb_mdb_ref["Symbol_Pos"]== "SELL"), 'SELL_EMA3'] = 0

                count = count + 1
```
